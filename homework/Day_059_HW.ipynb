### 用 digits (3個數字即可 , dataset.load_digits(n_class=3)) 資料嘗試跑 PCA


```python
from sklearn import datasets
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

digits = datasets.load_digits(n_class=3)
X = digits.data
y = digits.target
```

```python
X.shape
```

```python
plt.imshow(X[8,:].reshape(8,8),cmap = 'gray')
```

載入 相關套件 並 執行 PCA 實驗 ...

```python
pca = PCA(n_components = 3)
X_reduction = pca.fit_transform(X)
```

```python
X_reduction
```

```python
fig = plt.figure(1, figsize=(20, 10))
plt.clf()
ax = Axes3D(fig, rect=[0, 0, .95, 1], elev=60, azim=0)
plt.cla()

for name, label in [('zero', 0), ('one', 1), ('two', 2)]:
    ax.text3D(X[y == label, 0].mean(),
              X[y == label, 1].mean() + 1.5,
              X[y == label, 2].mean(), name,
              horizontalalignment='center',
              bbox=dict(alpha=.5, edgecolor='w', facecolor='w'))
# Reorder the labels to have colors matching the cluster results
y = np.choose(y, [1, 2, 0]).astype(np.float)

ax.scatter(X_reduction[:, 0], X_reduction[:, 1], X_reduction[:, 2], c=y, cmap=plt.cm.nipy_spectral,
           edgecolor='k',s = 80)

ax.w_xaxis.set_ticklabels([])
ax.w_yaxis.set_ticklabels([])
ax.w_zaxis.set_ticklabels([])

plt.show()
```
